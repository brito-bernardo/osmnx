```python

import osmnx as ox
import matplotlib.pyplot as plt
import networkx as nx

#Place name which is used to retrieve the data from OSM
```

```python
location_name = "Neópolis, Natal, Brazil"

#Retrieve the street networl

graph = ox.graph_from_place(location_name)

# Plot graph
fig, ax = ox.plot_graph(graph)
```

```python
place = {'city' : 'Porto', 'country' : 'Portugal'}

cf = '["highway"~"cycleway"]'

G = ox.graph_from_place(place, network_type="bike", custom_filter=cf, )

nc = ox.plot.get_node_colors_by_attr(G, attr="y", cmap="plasma")


fig, ax = ox.plot_graph(G, node_color=nc, save=True)
print(len(G), "cycleway")
```
