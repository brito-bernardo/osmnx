```python
import osmnx as ox
```

```python
place_name = "Porto, Portugal"

G = ox.graph_from_place(place_name, network_type = "drive")
```

```python
fig, ax = ox.plot_graph(G, node_color = "r")
```

```python
G_proj = ox.project_graph(G)
G2 = ox.consolidate_intersections(G_proj,rebuild_graph=True, tolerance=20, dead_ends=False)
fig, ax = ox.plot_graph(G2, node_color = "r")
```

```python
G3 = ox.graph_from_place(place_name, network_type = "all")
```

```python
fig, ax = ox.plot_graph(G3,node_color = 'none')
```

```python
place_name2 = "Maia, Portugal"
G4 = ox.graph_from_place(place_name2, network_type = "drive")
```

```python
place_name3 = "Matosinhos, Portugal"
G5 = ox.graph_from_place(place_name3, network_type = "drive")
```

```python
import networkx as nx
```

```python
G_combined = nx.compose(G, G4)
```

```python
G_combined = nx.compose(G_combined, G5)
```

```python
ox.plot_graph(ox.project_graph(G_combined), node_color = "none")
```
