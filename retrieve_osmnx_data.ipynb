```python
import osmnx as ox
```

```python
#Specify the name that is used to search for the data

place_name = "Paranhos ,Porto , Portugal"

#Fetch OSM street network from the Location

graph = ox.graph_from_place(place_name, network_type="drive")
```

```python
#Data type of the graph
type(graph)
```

```python
#Plot the streets

ox.plot_graph(graph)
```

```python
#Place polygon

area = ox.geocode_to_gdf(place_name)
```

```python
type(area)
```

```python
area
```

```python
area.plot()
```

```python
#Building footprints

buildings = ox.geometries_from_place(place_name, tags={'building':True})
```

```python
len(buildings)
```

```python
buildings.head(3)
```

```python
len(buildings.columns)
```

```python
buildings.columns
```

```python
#Retrieve restaurants

restaurants = ox.features.features_from_place(place_name ,tags = {'amenity': 'restaurant' })
```

```python
len(restaurants)
```

```python
restaurants.columns
```

```python
#Select some useful cols

cols = ['name', 'opening_hours', 'addr:city', 'addr:country', 'addr:housenumber', 'addr:postcode', 'addr:street',]

#Print only selected cols
restaurants[cols].head()
```

```python
#Retrieve nodes and edges

nodes, edges = ox.graph_to_gdfs(graph)
```

```python
nodes.head()
```

```python
edges.head()
```

```python
#Plotting the data

import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12,8))

#Plot the footprint
area.plot(ax = ax, facecolor = "black")

#Plot street edges

edges.plot(ax=ax, linewidth= 1)

#Plot buildings

buildings.plot(ax=ax, facecolor = 'khaki', alpha=0.7)

#Plot restaurants

restaurants.plot(ax=ax, color= 'red')
```

```python
#Park polygons

leisure = ox.geometries_from_place(place_name, tags = {'leisure': True})
```

```python
leisure.head(3)
```

```python
leisure["leisure"].value_counts()
```

```python
parks = leisure[leisure["leisure"].isin(["park", "playgound"])]
```

```python
parks.plot(color = "green")
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12,8))

#Plot the footprint
area.plot(ax = ax, facecolor = "black")

#Plot street edges

edges.plot(ax=ax, linewidth= 1)

#Plot buildings

buildings.plot(ax=ax, facecolor = 'khaki', alpha=0.7)

#Plot restaurants

restaurants.plot(ax=ax, color= 'red')

#Plot parks

parks.plot(ax = ax, color = "green")
plt.tight_layout()

```
