```python
import osmnx as ox
import networkx as nx
```

```python
#getting place

place_name = "Porto, Portugal"

G = ox.graph_from_place(place_name, network_type = 'drive')
```

```python
#All edge intersections
fig, ax = ox.plot_graph(G, node_color = "r")
```

```python
#consolidated intersections
G_proj = ox.project_graph(G)
ints = ox.consolidate_intersections(G_proj, rebuild_graph=False, tolerance=10,dead_ends=False)
len(ints)
```

```python
#
G2 = ox.consolidate_intersections(G_proj, rebuild_graph=True,tolerance=10 ,dead_ends=False)
len(G2)
```

```python
fig, ax = ox.plot_graph(G2, node_color="r")
```
