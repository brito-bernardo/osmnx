```python
#Import all the libraries

import osmnx as ox
import matplotlib.pyplot as plt
import networkx as nx
import geopandas as gpd
import pandas as pd
```

```python
#Fetch the OSM data from Porto

place_name = 'Paranhos ,Porto, Portugal'
graph = ox.graph_from_place(place_name)

#Project the graph

graph_proj = ox.project_graph(graph)

#Plot the graph

fig, ax = ox.plot_graph(graph_proj)




```

```python
#Calculate basic statistics of the street network

stats = ox.basic_stats(graph_proj)


#Calculate the coverage area of the street network

edges = ox.graph_to_gdfs(graph_proj, nodes=False, edges = True)
area = edges.unary_union.convex_hull.area


#Retrive the missing statistics using the area information
stats = ox.basic_stats(graph_proj, area=area)
print(stats)

```
