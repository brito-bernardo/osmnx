```python
#Import all the libraries

import osmnx as ox
import matplotlib.pyplot as plt
import networkx as nx
import geopandas as gpd
import pandas as pd
```

```python
# Area of interest
place_name = "Porto, Portugal"
graph = ox.graph_from_place(place_name, network_type="all")

# Plot the map of Porto
ox.plot_graph(ox.project_graph(graph), node_size=0, show=False)

```

```python
custom_filter = '["highway"="cycleway"]'
graph = ox.graph_from_place(place_name, network_type="bike", )

# Cycleway
ox.plot_graph(ox.project_graph(graph), edge_color='yellow', edge_linewidth=2, show=True)
```

```python
graph = ox.graph_from_place(place_name, network_type="all")
fig, ax = ox.plot_graph(ox.project_graph(graph), node_size=0, show=False)

ciclovias = ox.graph_from_place(place_name, network_type="bike")
ox.plot_graph(ox.project_graph(ciclovias), ax=ax, edge_color='yellow', edge_linewidth=2, show=True)

plt.show()
```

```python
cf = '["highway"~"motorway|primary"]'
G = ox.graph_from_place(place_name, network_type="drive", custom_filter=cf)
```

```python
fig, ax = ox.plot_graph(G, node_size=0)
```
