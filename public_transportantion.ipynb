```python
import folium
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point, Polygon
```

```python
place_name = "Porto, Portugal"

# Get bicycle parking data
bicycle_parking_tags = {'amenity': 'bicycle_parking'}

bicycle_parkings = ox.geometries_from_place(place_name, tags=bicycle_parking_tags)

#Get metro station parking data

metro_stations_tags = {'railway': 'subway_entrance'}
metro_stations = ox.geometries_from_place(place_name, tags=metro_stations_tags)

#Get the bus station data

bus_stops_tags = {'highway': 'bus_stop'}
bus_stops = ox.geometries_from_place(place_name, tags=bus_stops_tags)


map = folium.Map(location=[41.1579, -8.6291], zoom_start=13)

for _, row in bicycle_parkings.iterrows():
    # Check if the geometry is a Point
    if isinstance(row.geometry, Point):
        location = [row.geometry.y, row.geometry.x]
    # If it's a Polygon, use the centroid
    elif isinstance(row.geometry, Polygon):
        centroid = row.geometry.centroid
        location = [centroid.y, centroid.x]
    else:
        continue  # Skip if the geometry type is neither Point nor Polygon

    folium.Marker(
        location=location,
        icon=folium.Icon(icon='bicycle', color='blue'),
        tooltip='Bicycle Parking'
    ).add_to(map)



for _, row in metro_stations.iterrows():
    # Check if the geometry is a Point
    if isinstance(row.geometry, Point):
        location = [row.geometry.y, row.geometry.x]
    # If it's a Polygon, use the centroid
    elif isinstance(row.geometry, Polygon):
        centroid = row.geometry.centroid
        location = [centroid.y, centroid.x]
    else:
        continue  # Skip if the geometry type is neither Point nor Polygon

    folium.Marker(
        location=location,
        icon=folium.Icon(icon='subway', color='red'),
        tooltip='Metro station'
    ).add_to(map)


for _, row in bus_stops.iterrows():
    # Check if the geometry is a Point
    if isinstance(row.geometry, Point):
        location = [row.geometry.y, row.geometry.x]
    # If it's a Polygon, use the centroid
    elif isinstance(row.geometry, Polygon):
        centroid = row.geometry.centroid
        location = [centroid.y, centroid.x]
    else:
        continue  # Skip if the geometry type is neither Point nor Polygon

    folium.Marker(
        location=location,
        icon=folium.Icon(icon='bus', color='green'),
        tooltip='Bus stop'
    ).add_to(map)




map
```
